### 2.THE RELATION BETWEEN COMPUTATIONAL PHYSICS AND THE SOLVING OF THE TIME INDEPENDENT SCHRODINGER EQUATION (TISE):

The Schrödinger equation, the fundamental equation of quantum mechanics that defines the behavior of quantum systems, is solved using computational physics. The Schrödinger equation is a partial differential equation that is notoriously difficult to solve analytically for most systems, although it can be solved numerically using computer methods.
The infinite square well is a simple quantum system that is frequently used to illustrate computational methods. The infinite square well is a one-dimensional potential well with an infinite potential barrier at its edges and a constant potential inside the well. The Schrödinger equation for this system may be solved analytically, and the answer is a set of energy eigenstates with a discrete spectrum of permitted energies.
Computational physics can be used to solve the Schrödinger equation for more complex systems, such as a particle in a potential well that is not solvable analytically. In this case, the Schrödinger equation can be solved numerically using methods such as finite difference methods, finite element methods, or spectral methods, among others. These methods involve discretizing the space and time variables in the Schrödinger equation and solving the resulting matrix equations using numerical linear algebra techniques.
Computational methods can be used to compute the energy eigenstates and corresponding energies of the system under varied boundary conditions, such as varying well widths or potential depths, in the case of the infinite square well. This enables researchers to investigate how the system's energy spectrum varies as the boundary conditions are changed, and to compare the results to analytical solutions if available. Overall, computational physics is a necessary tool for solving the Schrödinger equation and analyzing the behavior of quantum systems such as the infinite square well.

 ## 1. The shooting Method: 
 The shooting technique in numerical analysis is a method for solving a boundary value problem by reducing it to an initial value problem. It entails solving the initial value issue for various initial conditions until one discovers a solution that also meets the boundary conditions of the boundary value problem. In layman's words, one "shoots" out trajectories from one boundary until one discovers the trajectory that "hits" the other boundary condition.

## 2. Finite-diffrence method: 
Finite-difference methods (FDM) are a class of numerical techniques used in numerical analysis to solve differential equations by approximating derivatives with finite differences. The spatial domain and time interval (if appropriate) are both discretized, or broken down into a finite number of steps, and the value of the solution at these discrete locations is approximated by solving algebraic equations with finite differences and values from neighboring points.Finite difference methods transform nonlinear ordinary differential equations (ODE) or partial differential equations (PDE) into a system of linear equations that may be solved using matrix algebra techniques. Because modern computers can perform these linear algebra computations efficiently, and because they are relatively simple to implement, FDM is widely used in modern numerical analysis. FDM, along with finite element methods, are now two of the most used methodologies to numerically solving PDEs.


```python
import numpy as np
import matplotlib.pyplot as plt

class Differentiation:
    def __init__(self, N):
        self.N = N
        self.a = 2 * np.pi
        self.x = np.linspace(0, 2 * np.pi, N)
        self.h = self.x[1] - self.x[0]
    
    def first_derivative(self, y):
        # Compute the matrix using the np.diag(np.ones(N),0) which creates a 
        # diagonal matrix of 1 of NxN size. Multiply by -1 to get -1 diagonal array.
        # You get an +1 off-diagonal array of ones, with np.diag(np.ones(N-1),1)
        # Note that you need N-1 for an NxN array, since the off diagonal is one smaller.
        # Add the two together and normalize by 1/h
        M = 1. / self.h * (np.diag(-1. * np.ones(self.N), 0) + np.diag(np.ones(self.N - 1), 1))
        
        # Compute the derivative of y into yp by matrix multiplication:
        yp = M.dot(y)
        
        # Print the first derivative of y
        print("First Derivative:")
        print(yp)
        
        return yp
    
    def second_derivative(self, y):
        # Compute the second derivative matrix
        Mdd = 1. / (self.h * self.h) * (np.diag(np.ones(self.N - 1), -1) + np.diag(-2. * np.ones(self.N), 0) + np.diag(np.ones(self.N - 1), 1))
        
        # Compute the second derivative of y into ypp by matrix multiplication:
        ypp = Mdd.dot(y)
        
        return ypp

# Define the number of points in our space, N
N = 128

# Compute the function y = sin(x). With numpy this is easy:
y = np.sin(np.linspace(0, 2 * np.pi, N))

# Create differentiation object
diff = Differentiation(N)

# Compute first derivative of y
yp = diff.first_derivative(y)

# Compute second derivative of y
ypp = diff.second_derivative(y)

# Plot the results.
plt.figure(figsize=(10, 7))
plt.plot(diff.x, y, label='y')
plt.plot(diff.x[:-1], yp[:-1], label='y`')  # Last value is invalid, don't plot
plt.plot(diff.x[1:-1], ypp[1:-1], label='y``')  # First and last value is invalid.
plt.legend()
plt.show()
```

### Infinite Square Well
The infinite square well is a quantum mechanics theoretical model in which a particle is contained within a one-dimensional box with unlimited potential walls on either side. The particle's energy is quantized, which means it can only have discrete energy levels, and its wave function is a standing wave with nodes at the well's walls. The model is frequently used to teach quantum physics and to investigate the behavior of confined particles.The very simplest system to solve for is the infinite square well, for which $V=0$. We will readily recognize the results as alternating $\cos(x)$ and $\sin(x)$ functions, and the energy levels are:
$$
E_i = \frac{n^2\pi^2\hbar^2}{2ma^2}
$$
First, we need to discuss a subtlety. The Infinite Square Well from $-a/2$ to $a/2$ has $V=\infty$ *at* these points. We get into trouble trying to entery $\infty$ in our potential, so what we need to do is just limit the coputational space from $-a/2+h$ to $a/2-h$, where $h$ is our step size. That way we force the wavefunction to zero at the end points. 
We compute this in the next box. I create $x_{full}$ as the full x-axis from $-a/2$ to $a/2$, but take $N+2$ steps. I then leave out the first and last point when calculating the wavefunctions. At the end, before plotting, I add a zero to the beginning and end of the wavefunctions, so that we get the expected result for plotting.


V(x) = { 0         , for 0 $\leq$ x $\leq$ a <br />
 &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;       V$_0$       , for a $\lt$ x $\leq$ b <br />
 &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;       infinity  , for x $\lt$ 0 or x $\gt$ b <br />
      } <br />

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg as scl
hbar=1
m=1
N = 512
a = 1.0
x = np.linspace(-a/2.,a/2.,N)
# We want to store step size, this is the reliable way:
h = x[1]-x[0] # Should be equal to 2*np.pi/(N-1)
V = 0.*x
Mdd = 1./(h*h)*(np.diag(np.ones(N-1),-1) -2* np.diag(np.ones(N),0) + np.diag(np.ones(N-1),1))
H = -(hbar*hbar)/(2.0*m)*Mdd + np.diag(V) 
E,psiT = np.linalg.eigh(H) # This computes the eigen values and eigenvectors
psi = np.transpose(psiT)   # We take the transpose of psiT to the wavefunction vectors can accessed as psi[n]
#
```
